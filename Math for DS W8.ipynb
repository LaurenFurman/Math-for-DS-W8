```python
import numpy as np
import matplotlib.pyplot as plt

T = np.array([
    [3, 6],
    [2, 4],
])

#define input space as a grid of points and coordinate matrices from coordinate vectors
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
xx, yy = np.meshgrid(x, y)

#apply the transformation to each point in the input space
xy = np.array([xx.flatten(), yy.flatten()])
trans = T @ xy

#plot the original points and their transformed locations
f, axes = plt.subplots(1, 2, figsize=(6, 3))

#create scatter plot with original points
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[0].set_title('Original points')

#add transformed points
axes[1].scatter(trans[0].reshape(xx.shape), trans[1].reshape(yy.shape), s=10, c=xx+yy)
axes[1].set_title('Transformed points')

#generate output
plt.show()
```

```python
#the transformation T scales and skews the vectors in the 2D space. 
#how much the vectors are scaled and skewed depends on the specific values in the transformation matrix T.
```
